```python
data = pd.DataFrame(np.array(result), columns=['review'])
data = data.join(pd.DataFrame(data.pop('review').tolist()))
```

```python
reset = data['score'].value_counts().reset_index()
reset.columns = ['score', 'Counts']
reset = reset.sort_values(by='score')
reset
```

```python
df = pd.DataFrame(data[['userName', 'content', 'score']])
```

<br>Simpan hasilnya dalam bentuk csv/excel.
<br> Tampilkan 100 kata slang yang Anda dapatkan dan tampilkan dalam bentuk Gunakan (bobot:25)

```python
import pandas as pd

slangwords_df = pd.read_csv('stopwords.csv')

def find_slang(text):
    slangwords = list(slangwords_df['Slangwords'])
    found_slang = []
    for word in text.split():
        if word in slangwords:
            found_slang.append(word)
    return found_slang

df['content'] = df['content'].apply(find_slang)

print(df)
```
